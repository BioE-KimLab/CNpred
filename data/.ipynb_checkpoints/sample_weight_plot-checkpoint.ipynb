```python
import rdkit
#Maybe not a perfect way to classify ethers/esters
#It cannot find cyclic ethers/esters
#There can be some errors in If/else statements

Ester_SMARTS = rdkit.Chem.MolFromSmarts('[CX3;!R](=[OX1])OC')
Ether_SMARTS = rdkit.Chem.MolFromSmarts('[OD2;!R]([#6])[#6]')
#Ether_SMARTS = rdkit.Chem.MolFromSmarts('*[#6][OD2][#6]*')
COOH_SMARTS = rdkit.Chem.MolFromSmarts('*-C(=O)[O;D1]')

def ether_or_ester(smiles):
    mol = rdkit.Chem.MolFromSmiles(smiles)
    Ester_count = len(mol.GetSubstructMatches(Ester_SMARTS))
    Ether_count = len(mol.GetSubstructMatches(Ether_SMARTS))
    COOH_count = len(mol.GetSubstructMatches(COOH_SMARTS))
    
    #Ether_count = Ether_count - Ester_count
    
    #print(Ester_count, Ether_count, COOH_count)
    
    #if Ester_count == COOH_count:
    #    return "neither"
    
    if Ester_count >= 1:
        if Ether_count >= 1:
            return "ester and ether"
        else:
            return "ester"
    else:
        if Ether_count >= 1:
            return "ether"
        else:
            return "neither"
```

```python
df['ester_or_ether'] = pd.Series([ether_or_ester(smiles) for smiles in df['SMILES']])
```

```python
import matplotlib
import numpy as np
#ax.set_xlim([245,850])
#ax.set_ylim([0,105])
#ax.set_xticks(np.arange(250,851,150))
#ax.set_yticks(np.arange(20,101,20))
matplotlib.rcParams.update({'errorbar.capsize': 3})
plt.rcParams["font.family"] = 'Arial'
plt.rcParams.update({'font.size': 20})


fig, ax1 = plt.subplots()
fig.set_figheight(6)
fig.set_figwidth(6)
color = 'tab:red'
ax1.set_xlabel('Weight(Tier 2, 3)/Weight(Tier 1)')
ax1.set_ylabel('Mean absolute error \n(Validation + Test)', color=color)
#ax1.scatter(df2['sample_weight'],df2['val_test_loss'], label='validation+test MAE', color = color)
#ax1.plot(df2['sample_weight'],df2['val_test_loss'], color = color)
ax1.scatter(df2['sample_weight'],df2['val_loss'], label='validation+test MAE', color = color)
ax1.plot(df2['sample_weight'],df2['val_loss'], color = color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticks(np.arange(0.1,1.01,0.1))
ax1.set_yticks([5.4,5.6,5.8,6.0,6.2])

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Mean absolute error \n (Training)', color=color)  # we already handled the x-label with ax1
ax2.scatter(df2['sample_weight'],df2['train_loss'], label='train MAE', color = color)
ax2.plot(df2['sample_weight'],df2['train_loss'], color = color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yticks([1.5,2.0,2.5,3.0,3.5, 4.0])

#fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
